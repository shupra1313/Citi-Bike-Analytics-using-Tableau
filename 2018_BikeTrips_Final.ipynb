```python
import pandas as pd
import os
import glob
```

```python
os.listdir('2018_Citi_Bike_data')
```

```python
path = "2018_Citi_Bike_data"
filelist = glob.glob(path + "/*.csv")
filelist
```

```python
list = []
for f in filelist:
    df = (pd.read_csv(f))
    list.append(df)
```

```python
concat_list = pd.concat(list, axis = 0, ignore_index = True)
concat_list
```

```python
concat_list.count()
```

```python
concat_list.to_csv("clean_data_2018.csv")
```
